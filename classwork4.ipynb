# Lexicon Access and Display

```python
from sklearn import preprocessing # Import preprocessing functions from scikit-learn 
import nltk # Import the Natural Language Toolkit library
nltk.download('opinion_lexicon')
from nltk.corpus import opinion_lexicon
from nltk.tokenize import word_tokenize

print('Total number of words in opinion lexicon', len(opinion_lexicon.words()))
print('Examples of positive words in opinion lexicon', # Print the first 10 positive words in the lexicon
      opinion_lexicon.positive()[:10])
print('Examples of negative words in opinion lexicon', # Print the first 10 negative words in the lexicon
      opinion_lexicon.negative()[:10])
```

# Creating a dictionary

```python
# Let's create a dictionary which we can use for scoring our review text
nltk.download('punkt')
df.rename(columns={"reviewText": "text"}, inplace=True)
pos_score = 1
neg_score = -1
word_dict = {}
 
# Adding the positive words to the dictionary
for word in opinion_lexicon.positive():
        word_dict[word] = pos_score
      
# Adding the negative words to the dictionary
for word in opinion_lexicon.negative():
        word_dict[word] = neg_score
```

# Binglu Score

```python
def bing_liu_score(text):
    sentiment_score = 0 # Initialize sentiment score to 0
    bag_of_words = word_tokenize(text.lower())
    for word in bag_of_words: # Iterate through each word in the text
        if word in word_dict:
            sentiment_score += word_dict[word]
    return sentiment_score  # Return the final sentiment score
```

```python
df['text'].fillna('no review', inplace=True)
df['Bing_Liu_Score'] = df['text'].apply(bing_liu_score)
```
