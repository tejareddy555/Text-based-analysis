```
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Download NLTK stopwords
nltk.download('stopwords')
nltk.download('punkt')

# Define the cars dataset as a list of dictionaries
car_data = [
    {'ID': 1, 'Make': 'Ford', 'Model': 'F-150', 'Year': 2018, 'Price': 35000, 'Mileage': 25000, 'Review': 'Reliable truck with good performance both on and off-road.'},
    {'ID': 2, 'Make': 'Toyota', 'Model': 'Camry', 'Year': 2019, 'Price': 28000, 'Mileage': 30000, 'Review': 'A comfortable and fuel-efficient sedan suitable for daily commuting.'},
    {'ID': 3, 'Make': 'Honda', 'Model': 'Accord', 'Year': 2017, 'Price': 25000, 'Mileage': 35000, 'Review': 'A reliable and practical midsize sedan with a spacious interior.'},
    {'ID': 4, 'Make': 'Chevrolet', 'Model': 'Silverado', 'Year': 2020, 'Price': 38000, 'Mileage': 20000, 'Review': 'A powerful and capable full-size pickup truck with modern features.'},
    {'ID': 5, 'Make': 'Toyota', 'Model': 'Rav4', 'Year': 2018, 'Price': 30000, 'Mileage': 28000, 'Review': 'A popular compact SUV known for its reliability and fuel efficiency.'},
    {'ID': 6, 'Make': 'Ford', 'Model': 'Explorer', 'Year': 2019, 'Price': 32000, 'Mileage': 32000, 'Review': 'A versatile and spacious SUV with comfortable seating for seven passengers.'},
    {'ID': 7, 'Make': 'Honda', 'Model': 'Civic', 'Year': 2018, 'Price': 23000, 'Mileage': 40000, 'Review': 'A fuel-efficient and fun-to-drive compact car with a well-designed interior.'},
    {'ID': 8, 'Make': 'Jeep', 'Model': 'Wrangler', 'Year': 2016, 'Price': 27000, 'Mileage': 45000, 'Review': 'An iconic off-road SUV with rugged styling and excellent off-road capabilities.'},
    {'ID': 9, 'Make': 'Nissan', 'Model': 'Altima', 'Year': 2020, 'Price': 29000, 'Mileage': 22000, 'Review': 'A comfortable and stylish midsize sedan with good fuel efficiency and safety features.'},
    {'ID': 10, 'Make': 'Subaru', 'Model': 'Outback', 'Year': 2017, 'Price': 26000, 'Mileage': 38000, 'Review': 'A reliable and versatile wagon with standard all-wheel drive and a spacious cargo area.'}
]

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(car_data)

# Define stop words
stop_words = set(stopwords.words('english'))

# Define function to remove stop words
def remove_stopwords(text):
    tokens = word_tokenize(text)
    filtered_text = [word for word in tokens if word.lower() not in stop_words]
    return ' '.join(filtered_text)

# Define function for word frequency analysis
def word_frequency(text):
    tokens = word_tokenize(text)
    freq_dist = nltk.FreqDist(tokens)
    return freq_dist

# Define function to generate word cloud
def generate_wordcloud(text):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

# Define pipeline
def text_processing_pipeline(text):
    text = remove_stopwords(text)
    freq_dist = word_frequency(text)
    generate_wordcloud(text)
    return freq_dist

# Apply pipeline to car reviews
for index, row in df.iterrows():
    print("Car:", row['Make'], row['Model'])
    print("Review:", row['Review'])
    freq_dist = text_processing_pipeline(row['Review'])
    print("Word Frequencies:", freq_dist)
    print("="*50)
```

```
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download the VADER lexicon for sentiment analysis
nltk.download('vader_lexicon')

# Define the dataset
dataset = [
    "I recently purchased this car and I'm extremely satisfied with its performance. It handles well on the road and has a sleek design.",
    "The fuel efficiency of this vehicle is impressive. I can go for miles without needing to refuel, which is great for long trips.",
    "I've had nothing but problems with this car. It constantly breaks down, and the maintenance costs are astronomical. Stay away from it!",
    "This car has all the features I was looking for. The interior is spacious, the infotainment system is intuitive, and the safety features are top-notch.",
    "The handling of this car is subpar. It feels sluggish on the road, and I'm not confident in its ability to handle sharp turns.",
    "I'm impressed with the acceleration of this vehicle. It has a powerful engine that provides a thrilling driving experience.",
    "The build quality of this car is excellent. It feels solid and well-built, which gives me confidence in its durability and reliability.",
    "The design of this car is uninspiring. It looks like every other car on the road, and there's nothing distinctive about it.",
    "I've owned this car for several years now, and it's still running smoothly. It's been a reliable companion for my daily commute.",
    "The technology in this car is outdated. The touchscreen interface is slow and unresponsive, and there are limited connectivity options."
]

# Initialize the sentiment intensity analyzer
sid = SentimentIntensityAnalyzer()

# Perform sentiment analysis on each sentence in the dataset
for i, sentence in enumerate(dataset):
    # Analyze the sentiment of the sentence
    sentiment_scores = sid.polarity_scores(sentence)

    # Determine the sentiment label based on the compound score
    if sentiment_scores['compound'] >= 0.05:
        sentiment_label = 'Positive'
    elif sentiment_scores['compound'] <= -0.05:
        sentiment_label = 'Negative'
    else:
        sentiment_label = 'Neutral'

    # Print the sentiment analysis results
    print(f"Sentence {i+1}: {sentence}")
    print(f"Sentiment Score: {sentiment_scores['compound']}")
    print(f"Sentiment Label: {sentiment_label}")
    print("=" * 50)
```

```
# Define health-related dataset
health_reviews = [
    "I feel great after starting this new exercise routine. My energy levels are up, and I'm sleeping better at night.",
    "I've been experiencing some pain in my lower back lately. It's been affecting my daily activities and I'm concerned.",
    "The medication prescribed by my doctor has been very effective in managing my symptoms. I'm feeling much better now.",
    "I'm struggling with maintaining a healthy diet. I often give in to unhealthy cravings and feel guilty afterwards.",
    "The side effects of this treatment are unbearable. I'm experiencing nausea, headaches, and fatigue on a daily basis.",
    "I've been following this wellness program for a few weeks now, and I'm already seeing positive changes in my mood and overall well-being.",
    "I'm feeling stressed and overwhelmed with work. It's taking a toll on my mental and physical health.",
    "This new supplement has made a significant difference in my joint pain. I can now move more freely without discomfort.",
    "I'm disappointed with the lack of progress in my fitness journey. Despite my efforts, I haven't seen any improvements in my strength or endurance.",
    "I'm grateful for the support of my healthcare team. They've been instrumental in helping me manage my chronic condition."
]

# Define positive and negative word lists from Bing Liu lexicon
positive_words = set([
    "great", "effective", "better", "positive", "wellness", "significant", "grateful"
])

negative_words = set([
    "pain", "struggling", "unbearable", "stress", "overwhelmed", "disappointed", "guilty"
])

# Define function to calculate Bing Liu sentiment score for a given review
def bing_liu_sentiment(review):
    # Tokenize the review into words
    words = review.lower().split()

    # Count occurrences of positive and negative words
    positive_count = sum(1 for word in words if word in positive_words)
    negative_count = sum(1 for word in words if word in negative_words)

    # Calculate sentiment score based on word counts
    score = positive_count - negative_count

    return score

# Perform sentiment analysis using Bing Liu lexicon on each health review
for i, review in enumerate(health_reviews):
    # Calculate Bing Liu sentiment score for the review
    score = bing_liu_sentiment(review)

    # Determine sentiment label based on score
    if score > 0:
        sentiment_label = 'Positive'
    elif score < 0:
        sentiment_label = 'Negative'
    else:
        sentiment_label = 'Neutral'

    # Print sentiment analysis results
    print(f"Review {i+1}: {review}")
    print(f"Bing Liu Sentiment Score: {score}")
    print(f"Sentiment Label: {sentiment_label}")
    print("=" * 50)
```
