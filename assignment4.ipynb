# Editing Date

```python
netflix_df["date_added"] = pd.to_datetime(netflix_df['date_added'], format='mixed')

netflix_df['month_added']= netflix_df['date_added'].dt.month
netflix_df['month_name_added']=netflix_df['date_added'].dt.month_name()
netflix_df['year_added'] = netflix_df['date_added'].dt.year


netflix_df.head()
```

# Data Visualization

```python
sns.countplot(data=netflix_df, x='type', order=['Movie', 'TV Show'])
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Type')
```

```python

labels = ['Movie', 'TV show']
size = netflix_df['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(10,8))
sns.countplot(x='rating',hue='type',data=netflix_df)
plt.title('Relation between Type and Rating')
plt.show()
```

```python
netflix_df['rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.show()
```
